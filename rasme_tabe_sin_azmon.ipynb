```python
# نمودار f(x) = sin(1 + e^(-x^2)) را در بازه 0 تا 5 رسم کنید
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(1+np.exp(-x**2))

x=np.linspace(0,5,100)
y=f(x)
plt.plot(x,y)
```
